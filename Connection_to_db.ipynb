```python
#import dependencies
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
```

```python
#store csv into df
csv_file = "cleaned_data/police_killings_cleaned.csv"
killings_df = pd.read_csv(csv_file, encoding="ISO-8859-1")
killings_df.head()
```

```python
engine = create_engine('sqlite:///killings_db.sqlite')
conn = engine.connect()
```

```python
#check for tables
engine.table_names()
```

```python
#Use pandas to load csv converted DataFrame into database
killings_df.to_sql(name='police_killings', con=engine, if_exists='append', index=True)
```

```python
#Confirm data has been added by querying the table
pd.read_sql_query('select * from police_killings', con=engine).head()
```
