```python
#import dependencies
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
```

```python
#store csv into df
csv_file = "cleaned_data/police_killings_2015.csv"
killings_df = pd.read_csv(csv_file, encoding="ISO-8859-1")
killings_df.head()
```

```python
killings_df['raceethnicity'].replace({'Hispanic/Latino':'Hispanic','Asian/Pacific Islander':'Asian'}, inplace=True)
killings_df['armed'].replace({'Firearm':'Yes','Non-lethal firearm':'Yes','Other':'Yes','Knife':'Yes','Vehicle':'Yes',
                             'Unknown':'Yes','Disputed':'Yes'},inplace=True)
```

```python
killings_df.head()
```

```python
engine = create_engine('sqlite:///killings_db.sqlite')
conn = engine.connect()
```

```python
#check for tables
engine.table_names()
```

```python
#Use pandas to load csv converted DataFrame into database
killings_df.to_sql(name='police_killings_2015', con=engine, if_exists='replace', index=True)
```

```python
#Confirm data has been added by querying the table
pd.read_sql_query('select * from police_killings_2015', con=engine).head()
```
